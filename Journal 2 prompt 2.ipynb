```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1,100)
def sin(x):
    return np.sin(x)

def cos(x): 
    return np.cos(x)

    
```

```python
sin_x = sin(x)
cos_x = cos(x)

figure, axes = plt.subplots(1,2)


axes[0].plot(x,sin_x)
axes[0].set_xlabel("x")
axes[0].set_ylabel("sin(x)")
axes[0].set_title("sin(x) from 0 to 1")

axes[1].plot(x,cos_x)
axes[1].set_xlabel("x")
axes[1].set_ylabel("cos(x)")
axes[1].set_title("cos(x) 0,1")
```

```python
figure.savefig("demo.pdf", format='pdf')
```
